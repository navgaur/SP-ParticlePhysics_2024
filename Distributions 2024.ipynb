```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
path="/content/dielectron.csv"
df=pd.read_csv(path)
print(df)
```

```python
print(df.groupby(['Q1']).size())
```

```python
x1=df[df['E1']<400]
x2=df[df['E2']<400]
#x1=df['E1']
#x2=df['E2']
colors = ['blue','red']
labels= ['E1','E2']
plt.figure(figsize=(18,6))
n, bins, edges = plt.hist([x1['E1'],x2['E2']],bins=30,color=colors,
                          label=labels, alpha = 0.5, align = 'mid')
#plt.hist([x1,x2],bins=20,color=colors,label=labels, alpha = 0.5)
plt.yscale('log')
plt.xticks(bins)
plt.legend(loc = 'best')
plt.title('Histogram of Particle Energy')
plt.xlabel('Energy')
plt.ylabel('Freq')
plt.show()
```
